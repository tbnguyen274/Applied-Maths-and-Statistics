## **2. Giải thuật**
**a. Phép khử Gauss (Gaussian elimination):**
> Bước 1. Xác định cột trái nhất không chứa toàn số 0

> Bước 2. Đổi chỗ hai dòng, nếu cần thiết, để đưa số hạng khác 0 nào đó ở dưới về đầu cột nhận được ở bước 1.

> Bước 3. Với số hạng đầu cột nhận được từ bước 2 là a khác 0, nhân dòng 1 với 1/a để nhận được một số dẫn đầu (leading 1).

> Bước 4. Cộng một bội số thích hợp của dòng 1 với từng dòng dưới để biến các số hạng này về số 0.

> Bước 5.  Loại bỏ các dòng đã làm xong. Trở lại bước 1 cho ma trận con này.

**b. Phép thế ngược (back substitution):**

> Bước 1: Biện luận trên các dòng gồm toàn số 0 của ma trận các hệ số.
 Nếu có một dòng gồm toàn số 0 của A mà hệ số vế phải (tự do)
 tương ứng là một số khác 0 thì kết luận hệ vô nghiệm và giải
 thuật chấm dứt. Ngược lại, bỏ tất cả các dòng gồm toàn số 0
 của ma trận các hệ số cũng như của vectơ các hệ số vế phải (tự
 do) và chuyển qua bước 2

 > Bước 2: Phép thế ngược (back substitution). Xác định các ẩn cơ sở là
 các ẩn tương ứng với các số dẫn đầu. Các ẩn còn lại được gọi là
 ẩn tự do (free variable). Xét hai khả năng:
 - Không có ẩn tự do. Hệ có nghiệm duy nhất nhận được bằng
 cách giải từ dưới lên trên (backward).
 - Có ít nhất một ẩn tự do. Hệ có vô số nghiệm và nghiệm tổng
 quát nhận được bằng cách giải từ dưới lên tìm biểu thức của
 các ẩn dẫn đầu theo các tham số của ẩn tự do (tùy ý).

## **3. Hàm Gauss_elimination(A)**

```python
# Input: A, là ma trận mở rộng của hệ phương trình Ax = b.
# Output: ma trận có dạng bậc thang có được từ ma trận A.

def Gauss_elimination(A):
    n = len(A)
    m = len(A[0])

    for i in range(n):
        # Xác định cột đầu tiên không chứa toàn số 0
        curCol = i
        while curCol < m and all(A[j][curCol] == 0 for j in range(i, n)):
            curCol += 1

        # Nếu không tìm được cột thích hợp, bỏ qua
        if curCol == m:
            continue

        if A[i][curCol] == 0:
            for j in range(i + 1, n):
                if A[j][curCol] != 0:
                    # Đổi dòng i và dòng j
                    A[i], A[j] = A[j], A[i]
                    break
            
        # Nhân dòng i với 1 / A[i][curCol] để biến phần tử này thành 1 (leading 1)
        A[i] = [a / A[i][curCol] for a in A[i]]

        # Cộng một bội số thích hợp của dòng 1 với từng dòng dưới để biến các số hạng này về số 0
        for j in range(i + 1, n):
            factor = A[j][curCol]
            for k in range(curCol, m):
                A[j][k] -= factor * A[i][k]
        
        # Biến -0 thành 0
        for j in range(m):
            if abs(A[i][j]) < 1e-10:
                A[i][j] = 0.0
        
        # in ra ma trận sau mỗi bước
        print("Bước", i + 1)
        for row in A:
            print(row)
        print()
        
    return A
```

## **4. Hàm back_substitution(A)**

```python
# Input: A, là ma trận bậc thang.
# Output: nghiệm của hệ phương trình Ax = b. Dùng thư viện để giải trường hợp có nghiệm nhiều.

import sympy as sp
sp.init_printing()

def back_substitution(A):
    n = len(A)
    m = len(A[0])

    # TH1: Hệ phương trình vô nghiệm
    for i in range(n):
        if all(A[i][j] == 0 for j in range(m - 1)) and A[i][m - 1] != 0:
            return "Kết luận: Hệ phương trình vô nghiệm."
    
    # Bỏ qua các dòng toàn 0
    A = [row for row in A if any(row[j] != 0 for j in range(m))]

    # back substitution

    # Xác định các ẩn cơ sở là các ẩn tương ứng với các số dẫn đầu. Các ẩn còn lại ẩn tự do (free variable). 
    # Tính số lượng ẩn tự do
    num_free_var = m - n - 1
    print("Số lượng biến tự do:", num_free_var)
    print("Số lượng biến cơ sở:", m - 1 - num_free_var)

    if num_free_var == 0:
        # TH2: Hệ phương trình có nghiệm duy nhất

        # Không có ẩn tự do. Hệ có nghiệm duy nhất nhận được bằng cách giải từ dưới lên trên (backward).
        x = [0] * n
        for i in range(n - 1, -1, -1):
            x[i] = A[i][m - 1] - sum(A[i][j] * x[j] for j in range(i + 1, n))
        
        print("Kết luận: Hệ phương trình có nghiệm duy nhất:")
        return x    # Trả về vector nghiệm
    
    else:
        #  TH3: Hệ phương trình có vô số nghiệm

        # Có ít nhất một ẩn tự do. Hệ có vô số nghiệm và nghiệm tổng quát nhận được bằng cách giải từ dưới lên tìm biểu thức
        # của các ẩn dẫn đầu theo các tham số của ẩn tự do (tùy ý).

        x = [sp.Symbol(f'x{i + 1}') for i in range(m - 1)]  # Tạo biến x1, x2, ..., xm
        solution = {}   # Lưu nghiệm dưới dạng từ điển
        free_vars = set(x)  # Tập hợp các biến tự do, mặc định là tất cả các biến

        # Duyệt từ hàng cuối lên đầu (backward)
        for row in range(n - 1, -1, -1):
            col = 0
            while col < m - 1 and A[row][col] == 0:
                col += 1
            
            # Khi tìm thấy cột chứa leading 1, đây là biến cơ sở (không còn là biến tự do)
            free_vars.discard(x[col])

            # Khởi tạo nghiệm với hằng số phía bên phải
            root = A[row][m - 1]
            
            for j in range(col + 1, m - 1):
                coeff = A[row][j]   # Hệ số của biến x[j]
                term = solution.get(x[j], sp.Symbol(f"t_{j+1}"))  # Lấy nghiệm hoặc biến tự do
                
                if coeff != 0:
                    root -= coeff * term  # Tiến hành tìm x dựa trên các nghiệm đã biết

            solution[x[col]] = root  # Lưu nghiệm

        # Gán tham số cho các biến tự do
        for var in free_vars:
            var_num = str(var).strip("x")  # Lấy chỉ số của biến x
            solution[var] = sp.Symbol(f"t_{var_num}")

        print("Kết luận: Hệ phương trình có vô số nghiệm.")
        print("Nghiệm tổng quát:")

        # Chuyển kết quả thành vector dùng thư viện sympy
        result_vector = sp.Matrix([[solution[var]] for var in x])

        sp.pprint(result_vector)

        return solution
```

## **5. Kiểm tra kết quả**

```python
# Test

# Hệ phương trình 1 - Hệ phương trình có nghiệm duy nhất

A1 = [[1, -1, -2], 
      [2, -1, 1], 
      [1, 1, 1]]

B1 = [-3, 1, 4]

# Ma trận mở rộng của hệ phương trình
C1 = [A1[i] + [B1[i]] for i in range(len(A1))]
print("Ma trận mở rộng của hệ phương trình 1:")
for row in C1:
    print(row)

print("Biến đổi ma trận C1 thành ma trận bậc thang:")
result_1 = Gauss_elimination(C1)

# giải hệ phương trình
result_1 = back_substitution(result_1)
print(result_1)
print()

# Hệ phương trình 2 - Hệ phương trình vô nghiệm

A2 = [[1, 1, -2],
      [2, 3, 3],
      [5, 7, 4]]

B2 = [4, 3, 5]

# Ma trận mở rộng của hệ phương trình
C2 = [A2[i] + [B2[i]] for i in range(len(A2))]
print("Ma trận mở rộng của hệ phương trình 2:")
for row in C2:
    print(row)

print("Biến đổi ma trận C2 thành ma trận bậc thang:")
result_2 = Gauss_elimination(C2)

# giải hệ phương trình
result_2 = back_substitution(result_2)
print(result_2)
print()

# Hệ phương trình 3 - Hệ phương trình có vô số nghiệm

A3 = [[0, 0, -2, 0, 7],
      [2, 4, -10, 6, 12],
      [2, 4, -5, 6, -5]]

B3 = [12, 28, -1]

# Ma trận mở rộng của hệ phương trình
C3 = [A3[i] + [B3[i]] for i in range(len(A3))]
print("Ma trận mở rộng của hệ phương trình 3:")
for row in C3:
    print(row)

print("Biến đổi ma trận C3 thành ma trận bậc thang:")
result_3 = Gauss_elimination(C3)

# giải hệ phương trình
result_3 = back_substitution(result_3)
print(result_3)
```

## **6. Mở rộng: Sử dụng thư viện và so sánh kết quả**

Ta có thể sử dụng hàm `numpy.linalg.solve` trong thư viện NumPy để giải hệ phương trình tuyến tính khi có nghiệm duy nhất. Nếu hệ vô nghiệm hoặc có vô số nghiệm, hàm sẽ cho kết quả thiếu chính xác hoặc báo lỗi vì nó chỉ áp dụng cho ma trận vuông có định thức khác 0.

```python
import numpy as np

# Kiểm tra lại 3 ví dụ trên

# Hệ phương trình 1 - Hệ có nghiệm duy nhất
A1 = np.array([[1, -1, -2], 
               [2, -1, 1], 
               [1, 1, 1]])

B1 = np.array([-3, 1, 4])

try:
    solution_1 = np.linalg.solve(A1, B1)
    print("Nghiệm của hệ 1:", solution_1)
except np.linalg.LinAlgError:
    print("Hệ phương trình 1 không có nghiệm duy nhất")

# Hệ phương trình 2 - Hệ vô nghiệm
A2 = np.array([[1, 1, -2],
               [2, 3, 3],
               [5, 7, 4]])

B2 = np.array([4, 3, 5])

try:
    solution_2 = np.linalg.solve(A2, B2)
    print("Nghiệm của hệ 2:", solution_2)
except np.linalg.LinAlgError:
    print("Hệ phương trình 2 không có nghiệm duy nhất")

# Hệ phương trình 3 - Hệ có vô số nghiệm
A3 = np.array([[0, 0, -2, 0, 7],
               [2, 4, -10, 6, 12],
               [2, 4, -5, 6, -5]])

B3 = np.array([12, 28, -1])

try:
    solution_3 = np.linalg.solve(A3, B3)
    print("Nghiệm của hệ 3:", solution_3)
except np.linalg.LinAlgError:
    print("Hệ phương trình 3 không có nghiệm duy nhất")
```
